```python
#Single key is slways used 
student = {"name":"pradip","age":30,"name":24}
print(student)

```

```python
## Accessing Dictionary Elements 
student = {"name":"pradip","age":30,"grade":"A"}
print(student["name"])
print(student["age"])
```

```python
## Accessing using get() method
print(student.get("name"))
print(student.get("age"))
print(student.get("last_name","Not Availabe"))
```

```python
## Modifying Dictionary Elements
#Dictionaries are mutable 
student = {"name":"pradip","age":30,"grade":"A"}
student["age"] = 31
print(student)
```

```python
# Nested Dictionary
student = { 
    "studwent1": {"name":"pradip","age":30,"grade":"A"},
    "student2": {"name":"john","age":25,"grade":"B"}
}
print(student)
```

```python
## ACCESS nested disctionary elements
print(student["studwent1"]["name"])
print(student["student2"]["grade"])
```

```python
student.items()
```

```python
# Iterating over nested dictionary
for student_id, student_info in student.items():
    print(f"{student_id}:{student_info}")
    for key,value in student_info.items():
        print(f"  {key}:{value}")
```

```python
# Disctionary Comprehension
squares = {x: x*2 for x in range(1,6)}
print(squares)
```

```python
#Condition dictionary comprehension
evens= {x: x**2 for x in range(10) if x%2==0}
print(evens)
```

```python
## USE a dictionary to count the frequency of elements in list
numbers = [1,2,2,2,3,3,3,4,4,4,4]
frequency = {}
for number in numbers:
    if number in frequency:
        frequency[number] += 1
    else:
        frequency[number] = 1

print(frequency)
```

```python
# Merge 2 dictionaries into one

dict1 = {"a":1,"b":2}
dict2 = {"b":3,"c":4}
merged_dict = {**dict1,**dict2}
print(merged_dict)
```
