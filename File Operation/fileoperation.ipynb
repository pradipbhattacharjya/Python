```python
## Read  a whole File
with open('file.txt', 'r') as file:
    content = file.read()
    print(content)
```

```python
## Read a file line by line
with open('file.txt', 'r') as file:
    for line in file:
        print(line.strip())  # Remove leading/trailing whitespace
```

```python
#Writing a file(overwriting)
with open('file.txt', 'w') as file:
        file.write("This is an example of writing to a file.\n")
        file.write("This will overwrite the existing content.")
```

```python
## Write a file(without overwriting)
with open('file.txt', 'a') as file:
        file.write("APPend operation taking place!\n")
        file.write("This will add to the existing content.")
```

```python
#writing a list of lines to a file
lines=['First line\n', 'Second line\n', 'Third line\n']
with open('file.txt', 'a') as file:
    file.writelines(lines)
```

```python
#Binary file operations
data = b'\x00\x01\x02\x03\x04'
with open('file.bin', 'wb') as file:
    file.write(data)
```

```python
#Reading a binary file
with open('file.bin', 'rb') as file:
    binary_content = file.read()
    print(binary_content)
```

```python
## Read the content froma source text file and write to a destination text file

with open('file.txt','r') as source_file:
    content = source_file.read()

with open('destination.txt','w') as dest_file:
    dest_file.write(content)
```

```python
#Read a text file and count the number of lines, words, and characters.

def count_text_file(file_path):
    with open(file_path, 'r') as file:
        content = file.read()
        lines = content.splitlines()
        words = content.split()
        characters = len(content)
        
        print(f"Lines: {len(lines)}")
        print(f"Words: {len(words)}")
        print(f"Characters: {characters}")
```

```python
## Writing and then reading a file

with open('example.txt', 'w+') as file:
    file.write("Hello world\n")
    file.write("This is an new line\n")

    ##  Move the file cursor to the beginning
    file.seek(0)

    # Read the content of the file
    content = file.read()
    print(content)
```
