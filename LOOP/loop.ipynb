```python
# While loop
count = 0

while count < 5:
    print(count)
    count += 1
```

```python
count = 0

while count%2 ==0:
    print(count)
    count += 1
```

```python
## Loop control statements
#break statement
for i in range(1, 10):
    if i == 5:
        break
    print(i)
```

```python
##continue statement
for i in range(10):
    if i%2 == 0:
        continue
    print(i)
```

```python
##pass
for i in range(10):
    if i == 3:
        print("The number is ",i)
        pass
    print(i)
```

```python
## Nested loops

for i in range(3):
    for j in range(2):
        print(f"i: {i}and j: {j}")
```

```python
#Example - Calaculate the sum of first N natural numbers using a while and for loop
n = 10     
sum = 0
count = 1

while count <= n:
    sum += count
    count += 1
print(f"The sum of first {n} natural numbers is: {sum}")
```

```python
n = 10
sum = 0
for i in range(11):
    sum += i
print(f"The sum of first {n} natural numbers is: {sum}")
```

```python
#Exmple -Prime numbers between 1 to 100
for num in range(1, 101):
    if num > 1:
        for i in range(2, num):
            if (num % i) == 0:
                break
        else:
            print(num)
```
