```python
# if statement
age = 18

if age >= 18:
    print("You are an adult.")  
```

```python
#else statement
age = 16

if age >= 18:
    print("You are an adult.")
else:
    print("You are a minor.")
```

```python
## elif statement
age = 65
if age < 18:
    print("You are a minor.")
elif age < 65:
    print("You are an adult.")
else:
    print("You are a senior citizen.")
```
