```python
import pandas as pd
```

```python
# Load the Toyota sales dataset
toyota_sales_data = pd.read_csv("toyota_sales_data.csv")
```

```python
toyota_sales_data.head()
```

```python
toyota_sales_data["commission_pct"] = toyota_sales_data["commission_pct"].fillna(0)
```

```python
toyota_sales_data.head(10)
```

```python
toyota_sales_data["commission_amount"] = toyota_sales_data["sale_amount"] * toyota_sales_data["commission_pct"]
```

```python
toyota_sales_data.head(10)
```

```python
toyota_sales_data["sale_category"] = "Low"
```

```python
toyota_sales_data.loc[toyota_sales_data["sale_amount"] > 30000, "sale_category"] = "Medium"
```

```python
toyota_sales_data.head(20)
```

```python
toyota_sales_data = toyota_sales_data.drop(["sale_rep_id","sale_status"],axis=1)
```

```python
toyota_sales_data
```

```python
toyota_sales_data.loc[toyota_sales_data["sale_amount"] > 40000, "sale_category"] = "High"
```

```python
toyota_sales_data
```

```python
def get_sale_category(sale):
    if sale["sale_amount"] > 40000:
        return "High"
    elif sale["sale_amount"] > 30000:
        return "Medium"
    else:
        return "Low"
```

```python
toyota_sales_data["sale_category"] = toyota_sales_data.apply(get_sale_category, axis=1)
```
