```python
import pandas as pd
```

```python
#Load the CSV files
sales_reps_data = pd.read_csv("sales_reps_data.csv")
toyota_sales_data = pd.read_csv('toyota_sales_data.csv')
```

merged inner join

```python
merged_data = pd.merge(
    sales_reps_data,
    toyota_sales_data,
    left_on='rep_id',
    right_on='sale_rep_id',
    how = "inner"
)
```

```python
merged_data
```

```python
sales_by_rep = merged_data.\
    groupby(["rep_id","first_name","last_name","region"])["sale_amount"].\
    sum()
```
