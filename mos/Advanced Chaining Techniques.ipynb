```python
import pandas as pd
from babel.util import missing
```

```python
#Load the Toyota sales dataset
toyota_sales_data = pd.read_csv("toyota_sales_data.csv")
sales_reps_data = pd.read_csv("sales_reps_data.csv")
```

```python
missing_commission = toyota_sales_data["commission_pct"].isnull().sum()
```

```python
missing_commission
```

```python
toyota_sales_data.fillna({"commission_pct":0}) \
    .query("sale_status == 'Completed'") \
    .assign(commission_earned =lambda x: round(x["sale_amount"] * x["commission_pct"], 2))
```

```python
pd.merge(
    sales_reps_data,
    toyota_sales_data,
    left_on="rep_id",
    right_on="sale_rep_id",
    how="inner"

)   .fillna({"commission_pct":0}) \
    .query("sale_status == 'Completed'") \
    .assign(commission_earned =lambda x: round(x["sale_amount"] * x["commission_pct"], 2))

```

```python
pd.merge(
    sales_reps_data,
    toyota_sales_data,
    left_on="rep_id",
    right_on="sale_rep_id",
    how="inner"

)   .fillna({"commission_pct":0}) \
    .query("sale_status == 'Completed'") \
    .assign(commission_earned =lambda x: round(x["sale_amount"] * x["commission_pct"], 2))\
    .groupby("region").agg(
        total_sales = ("sale_amount", "sum"),
        avg_commission =("commission_earned", "mean")

    ) \
    .reset_index()
```
