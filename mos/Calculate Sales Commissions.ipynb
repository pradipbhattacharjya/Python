```python
import pandas as pd
```

```python
#load the Toyota sales dataset
toyota_sales_data = pd.read_csv("toyota_sales_data.csv")
sales_reps_data =pd.read_csv("sales_reps_data.csv")
```

```python
toyota_sales_data.head()
```

```python
sales_reps_data.head()
```

```python
pd.merge(
    sales_reps_data,
    toyota_sales_data,
    left_on="rep_id",
    right_on="sale_rep_id",
    how="inner"
).fillna({"commission_pct": 0}) \
    .assign(commission_earned=lambda x: round(x["sale_amount"] * x["commission_pct"]))\
    .assign(sale_month=lambda x: pd.to_datetime(x["sale_date"]).dt.to_period("M"))
```

```python
monthly_summary = pd.merge(
    sales_reps_data,
    toyota_sales_data,
    left_on="rep_id",
    right_on="sale_rep_id",
    how="inner"
).fillna({"commission_pct": 0}) \
    .assign(commission_earned=lambda x: round(x["sale_amount"] * x["commission_pct"]))\
    .assign(sale_month=lambda x: pd.to_datetime(x["sale_date"]).dt.to_period("M")) \
    .groupby(["sale_month", "rep_id", "first_name","last_name","email"]) \
    .agg(
        total_sale=("sale_amount", "sum"),
        total_commission=("commission_earned","sum")
    )\
    .reset_index()

```

```python
monthly_summary
```

```python
monthly_summary.to_csv("monthly_sales_commissions.csv",index=False)
```
