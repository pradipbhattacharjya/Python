```python
## Why functions?
num = 24
if num%2==0:
    print("the number is even")
else:
    print("the number is odd")
```

```python
def even_or_odd(num):
    """This function finds even or odd"""
    if num%2==0:
        print( "the number is even")
    else:
        print("the number is odd")
```

```python
# Call this function
even_or_odd(24)
```

```python
## function with multiple parameters
def add(a,b):
    """This function adds two numbers"""
    return a+b

result = add(5,10)
print(result)
```

```python
## Default parameters
def greet(name):
    """This function greets the user"""
    print(f"Hello, {name} Welcome to paradise")


greet("Pradip")
```

```python
def greet(name="Guest"):
    """This function greets the user"""
    print(f"Hello, {name} Welcome to paradise")

greet()
```

```python
## Variable Length Arguments
## Positional And Keyword Arguments
def print_numbers(*args):
    for num in args:
        print(num)
```

```python
print_numbers(1,2,3,4,5,"pradip")
```

```python
## Keyword Arguments
def print_details(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")
```

```python
print_details(name="Pradip", age=30, city="New York")
```

```python
def print_details(*args,**kwargs):
    for val in args:
        print(f"Positional argument: {val}")

        
    for key, value in kwargs.items():
        print(f"{key}: {value}")
```

```python
print_details(1,2,3,"suba",name="Pradip", age=30, city="New York") 
```
