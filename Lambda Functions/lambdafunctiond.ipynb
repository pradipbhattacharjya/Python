```python
def addition(a,b):
    return a + b
```

```python
addition(2,3)
```

```python
addition =lambda a,b:a+b
type(addition)
print(addition(5,6))
```

```python
def even(num):
    if num%2==0:
        return True
    
even(2)
```

```python
even1=lambda num:num%2==0
print(even1(12))
```

```python
def addition(x, y, z):
    return x + y +z

addition(2,3,4)
```

```python
addition1=lambda x,y,z:x+y+z
print(addition1(2,3,4))
```

```python
##map()- applies a function to all items in a list
numbers=[1,2,3,4,5,6]
def square(num):
    return num**2

square(2)
```

```python
list(map(lambda x:x**2, numbers))
```
